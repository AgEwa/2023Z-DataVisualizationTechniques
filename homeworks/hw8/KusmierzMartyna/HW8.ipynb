## Zadanie 1 - `numpy` + `pandas` (2 pkt)
Używając funkcji z pakietów `numpy` oraz `pandas` odpowiedz na poniższe pytania dotyczące zbioru danych o albumach muzycznych.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("rym_top5000.csv")
```

```python
df.head()
```

#### Jacy artyści z ponad 5 albumami na liście mają największą medianę ocen dla swoich albumów? Podaj top 10.

```python
df1 = df.groupby(by = "Artist Name").agg(Mediana = ("Average Rating", "median"),
                                        Albumy = ("Ranking", "count")).reset_index()

df1 = df1[df1.Albumy > 5].sort_values("Mediana", ascending = False).head(10).reset_index(drop=True)
df1

```

#### Jacy artyści mają największą różnicę między najgorszym i najlepszym średnim ratingiem dla swojego albumu? Podaj top 10  oraz najwyższą i najniższą ocenę dla każdego z tych artystów. 

```python
df2 = df.groupby(by = "Artist Name").agg(Najlepszy = ("Average Rating", "max"),
                                        Najgorszy = ("Average Rating", "min")).reset_index()
df2["Różnica"] = df2["Najlepszy"] - df2["Najgorszy"]
df2.sort_values("Różnica", ascending=False).head(10).reset_index(drop=True)
```

#### Podaj top 10 albumów o największym stosunku liczby ocen do liczby recenzji. Weź pod uwagę tylko albumy o przynajmniej 5000 ocenach i 15 recenzjach.

```python
df3 = df.loc[df["Number of Reviews"] > 15,]
df3 = df3.loc[df3["Number of Ratings"].str.replace(',','').astype(int) > 5000,]
df3["Stosunek Ocen"] = df3["Number of Ratings"].str.replace(',','').astype(int) / df3["Number of Reviews"]
df3 = df3.sort_values("Stosunek Ocen", ascending=False).head(10)[["Album","Stosunek Ocen"]].reset_index(drop=True)
df3
```

#### Z których lat pochodzi najwięcej albumów na liście najpopularniejszych wśród krytyków i najpopularniejszych wśród oceniających? 
a) Podaj 5 lat, w których wydano najwięcej albumów z listy mających liczbę recenzji większą niż Q3 (trzeci kwartyl).

b) Podaj 5 lat, w których wydano najwięcej albumów z listy mających liczbę ocen większą niż Q3 (trzeci kwartyl).


```python
import warnings
warnings.filterwarnings("ignore")

df4 = df[df["Number of Reviews"] > df["Number of Reviews"].quantile(0.75)]
df4["Rok"] = df4["Release Date"].str[-4:]
df4 = df4.groupby("Rok")["Rok"].agg(["count"])
df4 = df4.sort_values("count", ascending = False).head(5).reset_index()
df4
```

```python
df5 = df[df["Number of Ratings"] > df["Number of Ratings"].quantile(0.75)]
df5["Rok"] = df5["Release Date"].str[-4:]
df5 = df5.groupby("Rok")["Rok"].agg(["count"])
df5 = df5.sort_values("count",ascending = False).head(5).reset_index()
df5
```
